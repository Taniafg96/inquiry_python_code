# ***Decoding / Encoding in python***

## 1.- **ROT13**

Enconding / decoding example with ROT-13 ('rotate by 13 places')


![ROT-13 image](source/rot-13.png)


[Wikipedia Article ROT-13](https://en.wikipedia.org/wiki/ROT13)

```python
s = 'This is my first ROT13 excercise!'
```

```python
code = str.maketrans( 
        "ABCDEFGHIJKLMabcdefghijklmNOPQRSTUVWXYZnopqrstuvwxyz", 
        "NOPQRSTUVWXYZnopqrstuvwxyzABCDEFGHIJKLMabcdefghijklm")

print(f'Encoding/Decoding: {str.translate(s, code)}')
```

```python
import codecs

print(f'Encoding/Decoding: {codecs.decode(s, "rot13")}')
```
